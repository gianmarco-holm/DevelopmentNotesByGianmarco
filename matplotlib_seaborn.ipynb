2. **Permite hallazgos en nuestros datos**

También se utiliza en la ciencia de datos orientada hacia el machine learning o inteligencia artificial, como las líneas de entrenamiento. ### Librerías para visualización de datos

En la visualización de datos encontrarás dos librerías muy importantes para empezar:

* Matplotlib
    - Emula los comandos de Matlab
    - Usa Numpy
    - Escrita en Python
    - Es muy simple de escribir
    - Rápida
    - Personalizable

* Seaborn
    - Librería montada sobre Matplotlib

````python
import matplotlib.pyplot as plt
import numpy as np

edad = np.array([15, 16, 17, 18, 19])
frecuencia = np.array([3, 4, 5, 6, 2])

# Establecer un estilo de gráfico de barras más agradable
plt.style.use('seaborn-v0_8-darkgrid')

# Crear un gráfico de barras con colores personalizados
colores = ['skyblue' if f > max(frecuencia)/2 else 'lightcoral' for f in frecuencia]
fig, axes = plt.subplots(figsize=(4, 3))
bars = axes.bar(edad, frecuencia, color=colores)

# Etiquetas y título
axes.set_xlabel('Edad', fontsize=10)
axes.set_ylabel('Frecuencia', fontsize=10)
axes.set_title('Gráfico de Barras - Frecuencia por Edad', fontsize=12)

# Agregar etiquetas en la parte superior de las barras
for bar in bars:
    yval = bar.get_height()
    axes.text(bar.get_x() + bar.get_width()/2, yval + 0.1, round(yval, 1), ha='center', va='bottom', fontsize=8)

# Mostrar el gráfico
plt.show()
````

### Subplot

Subplot permite crear gráficos dentro de una gráfica. Esto lo hace a través de una matriz de gráficos y se puede acceder a ellos a través de índices:

#### Parámetros en Subplot
|Filas|Columnas|Index|
|--|--|--|

````python
# Crear fila y dos columnas
plt.subplot(1,2,1)
plt.plot(x,y,'r--')

plt.subplot(1,2,2)
plt.pie(y)

plt.show()
````

````python
# Agregar diferentes Plot al gráfico
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.plot(y,x,'b:')

plt.subplot(1,2,2)
plt.pie(y)

plt.show()
````

````python
# Invertir la matriz de los gráficos
plt.subplot(2,1,1)
plt.plot(x,y,'g>-')

plt.subplot(2,1,2)
plt.pie(x)

plt.show()
````

````python
plt.subplot(2,2,1)
plt.plot(x,y,'r:')

plt.subplot(2,2,2)
plt.hist(z)

plt.subplot(2,2,3)
plt.plot(y,z,'b-x')

plt.subplot(2,2,4)
plt.pie(z)

plt.show()
````

### Método orientado a objetos

Un objeto define una figura, esa figura es un lienzo en el cual podemos introducir diferentes grá ficas(axes), de las cuales cada una posee sus propios ejes(axis).

![axes](./images/axes.webp)

**Diferencias entre Subplot y Object Oriented**

| Característica     | Subplot                  | Orientado a Objetos     |
|--------------------|-------------------------|-------------------------|
| Rápido             | Menos personalización   | Mayor personalización  |
| Fácil              | Menos código            | Más código             |
| Una sola figura    | Conveniente            | Más amigable a múltiples diagramas |


````python
# creamos los arrays
x = np.linspace(0,5,11)
y = x ** 2
z = y * 3
````

**Parametros del axes:**

| Característica        | Pos. Eje x               | Pos. Eje y               | Size graph eje x          | Size graph eje y          |
|-----------------------|--------------------------|--------------------------|---------------------------|---------------------------|
|                      | Posición que va tener en el eje 'x'      | Posición que va tener en el 'y'       | Tamaño del gráfico en eje 'x' | Tamaño del gráfico en eje 'y' |


````python
# Utilizamos el concepto de objeto para crear una gráfica más personalizable
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.5,0.4])
axes.plot(x, y,'b')
plt.show()
````

````python
plt.figure().add_axes([0.1, 0.1, 0.5, 0.5]).plot(x,y, 'b')
````

````python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.6,0.6])
axes2 = fig.add_axes([0.75,0.1,0.15,0.15])
axes3 = fig.add_axes([0.75,0.3,0.15,0.15])
axes4 = fig.add_axes([0.75,0.5,0.15,0.15])

axes1.plot(x,y,'b')
axes2.hist(x)
axes3.pie(y)
axes4.plot(y,x,'g')
axes4.set_facecolor('gray')

plt.show()
````

````python
# Podemos crear varias gráficas dentro del lienzo llamando el método fig.
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.5,0.4])
axes2 = fig.add_axes([0.15,0.35,0.2,0.1])

axes.plot(x,y, 'b')
axes2.plot(y,x, 'r:')
axes2.set_facecolor('grey')
plt.show()
````

### Subplots

Con subplots se puede trabajar en un arreglo de gráficas a las cuales se accede a través de los índices.

````python
# Creamos los datos que vamos a utilizar
x = np.linspace(0,5,11)
y = x**2
z = y/2
````

````python
# Para crear una gráfica que sea tipo Fig. necesitas trabajar el siguiente código
fig, axes = plt.subplots(1,2)
axes[0].plot(x,y)
axes[1].pie(x)
plt.show()
````

````python
fig, (axes1, axes2) = plt.subplots(1,2)
axes1.scatter(x,y)
axes2.bar(x,y)
plt.show()
````

````python
fig, axes = plt.subplots(2,4)
axes[0,0].stem(x,y)
axes[0,1].step(x,y)
axes[0,2].pie(x)
axes[0,3].fill_between(y,x)
fig.tight_layout()
````

````python
# Alternativa para generar una matriz de tipo fig, accediendo a cada gráfica mediante los nombres dados en una tupla
fig, ((axes1, axes2, axes3, axes4), (axes5, axes6, axes7, axes8)) = plt.subplots(2,4)
axes1.plot(x, np.sin(x), 'b')
axes2.plot(y, np.cos(x), 'purple')
axes3.plot(x, np.tan(y), 'orange')
axes4.plot(y, np.cos(y)**2,'black')

#mejorar la visualización de las figuras
fig.tight_layout()
````

````python
fig, axes = plt.subplots(1,3)
axes[0].stem(z,x)
axes[1].step(x,y)
axes[2].fill_between(x,y, color='skyblue')
plt.tight_layout()
````

### Leyendas, etiquetas, títulos, tamaño

Para dar contexto a nuestros gráficos necesitamos usar títulos, leyendas, tamaño o etiquetas, para que nuestra gráfica tenga un contexto más amplio.

````python
fig, axes = plt.subplots(figsize=(5,3))# figsize: ajusta el tamaño de la figura
axes.plot(x,y, label='sin(x)')# Para usar legend tienes que agregar un label al trazo
axes.set_title('Relación x - y')# Se agrega un titulo
axes.set_xlabel('Eje X')# Se agrega un titulo al eje
axes.set_ylabel('Eje Y')
axes.legend(loc='upper left')# Se agrega una legenda, y se usa loc para ubicar la legenda
plt.show()
````

````python
# Tamaño de marcadores (markersize) y color (markerfacecolor)
fig, ax = plt.subplots(figsize = (4,4))
ax.plot(x,x+3,color = 'blue', linestyle = 'dashed', marker = '8', markersize = 10, markerfacecolor= "#37D842")

ax.plot(x,x+4, color = 'black',linestyle = '--', marker = 'P', markerfacecolor="#FF0000")
````

### Set

Set es un método de Seaborn que permite configurar el estilo, fuente y color de las gráficas.

````python
# Importamos librerias
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
````

````python
# Crear gráficas de Barplot
x = pd.Series(['A', 'B', 'C'])
y = pd.Series([1, 3, 2])
sns.barplot(x=x, y=y)
plt.show()
````

````python
# Configurar fuentes y estilos
sns.set(style='dark', palette='dark', font="Verdana", font_scale=1)
# style: estilo de la grilla, palette: estilo de la paleta, font: fuente de letra (tiene que estar con comillas dobles), font_scale: tamaño de letra
sns.barplot(x=x, y=y)
plt.show()
````

**¿Qué necesitas saber al usar Set?**

* Este parámetro afecta a todas las gráficas que creamos, incluidas las que no usan Seaborn.
* Podemos resetear los valores utilizando el método reset_orig.
* Este parámetro es un atajo de set_theme, que es el recomendado según la documentación

### Jointplot y Pairplot

En resumen, jointplot es ideal para explorar la relación entre dos variables, mientras que pairplot es más adecuado para explorar relaciones y distribuciones entre múltiples variables numéricas en un conjunto de datos. Ambas funciones son herramientas valiosas en el kit de herramientas de visualización de datos de Seaborn, y la elección entre ellas dependerá de los objetivos específicos de la visualización.

#### Jointplot

````python
# Joinplot une 2 distintas gráficas en una sola
sns.jointplot(x='total_bill', y='tip', data= tips, hue='sex');
````

### Heatmap

Heatmap es un tipo de gráfico enfocado a una estructura matricial. Heatmap correlaciona todas las variables numéricas del dataset.

````python
# Correlación entre variables
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

# Elimine las columnas no numéricas antes de calcular la correlación
numeric_tips = tips.select_dtypes(include=['float64', 'int64'])
numeric_tips.corr()
# Muestra las variables correlacionadas entre si
````

````python
# Heatmap
sns.heatmap(numeric_tips.corr());
````

````python
# Se pueden agregar diferentes parámetros. annot muestra el valor de la correlación,
# cmap color,linewidthsespacio entre variables, linecolor color de las líneas, vminv, max valores máximos y mínimos, cbar=False eliminar la barra
sns.heatmap(numeric_tips.corr(), annot= True, cmap='coolwarm', linewidths=5, linecolor='black', vmin=0.5,vmax=1,cbar=True);
````
