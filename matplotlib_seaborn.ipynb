### Librer칤as para visualizaci칩n de datos

En la visualizaci칩n de datos encontrar치s dos librer칤as muy importantes para empezar:

* Matplotlib
    - Emula los comandos de Matlab
    - Usa Numpy
    - Escrita en Python
    - Es muy simple de escribir
    - R치pida
    - Personalizable

* Seaborn
    - Librer칤a montada sobre Matplotlib

````python
import matplotlib.pyplot as plt
import numpy as np

edad = np.array([15, 16, 17, 18, 19])
frecuencia = np.array([3, 4, 5, 6, 2])

# Establecer un estilo de gr치fico de barras m치s agradable
plt.style.use('seaborn-v0_8-darkgrid')

# Crear un gr치fico de barras con colores personalizados
colores = ['skyblue' if f > max(frecuencia)/2 else 'lightcoral' for f in frecuencia]
fig, axes = plt.subplots(figsize=(4, 3))
bars = axes.bar(edad, frecuencia, color=colores)

# Etiquetas y t칤tulo
axes.set_xlabel('Edad', fontsize=10)
axes.set_ylabel('Frecuencia', fontsize=10)
axes.set_title('Gr치fico de Barras - Frecuencia por Edad', fontsize=12)

# Agregar etiquetas en la parte superior de las barras
for bar in bars:
    yval = bar.get_height()
    axes.text(bar.get_x() + bar.get_width()/2, yval + 0.1, round(yval, 1), ha='center', va='bottom', fontsize=8)

# Mostrar el gr치fico
plt.show()
````

### Subplot

Subplot permite crear gr치ficos dentro de una gr치fica. Esto lo hace a trav칠s de una matriz de gr치ficos y se puede acceder a ellos a trav칠s de 칤ndices:

#### Par치metros en Subplot
|Filas|Columnas|Index|
|--|--|--|

````python
# Crear fila y dos columnas
plt.subplot(1,2,1)
plt.plot(x,y,'r--')

plt.subplot(1,2,2)
plt.pie(y)

plt.show()
````

````python
# Agregar diferentes Plot al gr치fico
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.plot(y,x,'b:')

plt.subplot(1,2,2)
plt.pie(y)

plt.show()
````

````python
# Invertir la matriz de los gr치ficos
plt.subplot(2,1,1)
plt.plot(x,y,'g>-')

plt.subplot(2,1,2)
plt.pie(x)

plt.show()
````

````python
plt.subplot(2,2,1)
plt.plot(x,y,'r:')

plt.subplot(2,2,2)
plt.hist(z)

plt.subplot(2,2,3)
plt.plot(y,z,'b-x')

plt.subplot(2,2,4)
plt.pie(z)

plt.show()
````

### M칠todo orientado a objetos

Un objeto define una figura, esa figura es un lienzo en el cual podemos introducir diferentes gr치 ficas(axes), de las cuales cada una posee sus propios ejes(axis).

![axes](./images/axes.webp)

**Diferencias entre Subplot y Object Oriented**

| Caracter칤stica     | Subplot                  | Orientado a Objetos     |
|--------------------|-------------------------|-------------------------|
| R치pido             | Menos personalizaci칩n   | Mayor personalizaci칩n  |
| F치cil              | Menos c칩digo            | M치s c칩digo             |
| Una sola figura    | Conveniente            | M치s amigable a m칰ltiples diagramas |


````python
# creamos los arrays
x = np.linspace(0,5,11)
y = x ** 2
z = y * 3
````

**Parametros del axes:**

| Caracter칤stica        | Pos. Eje x               | Pos. Eje y               | Size graph eje x          | Size graph eje y          |
|-----------------------|--------------------------|--------------------------|---------------------------|---------------------------|
|                      | Posici칩n que va tener en el eje 'x'      | Posici칩n que va tener en el 'y'       | Tama침o del gr치fico en eje 'x' | Tama침o del gr치fico en eje 'y' |


````python
# Utilizamos el concepto de objeto para crear una gr치fica m치s personalizable
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.5,0.4])
axes.plot(x, y,'b')
plt.show()
````

````python
plt.figure().add_axes([0.1, 0.1, 0.5, 0.5]).plot(x,y, 'b')
````

````python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.6,0.6])
axes2 = fig.add_axes([0.75,0.1,0.15,0.15])
axes3 = fig.add_axes([0.75,0.3,0.15,0.15])
axes4 = fig.add_axes([0.75,0.5,0.15,0.15])

axes1.plot(x,y,'b')
axes2.hist(x)
axes3.pie(y)
axes4.plot(y,x,'g')
axes4.set_facecolor('gray')

plt.show()
````

````python
# Podemos crear varias gr치ficas dentro del lienzo llamando el m칠todo fig.
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.5,0.4])
axes2 = fig.add_axes([0.15,0.35,0.2,0.1])

axes.plot(x,y, 'b')
axes2.plot(y,x, 'r:')
axes2.set_facecolor('grey')
plt.show()
````

### Subplots

Con subplots se puede trabajar en un arreglo de gr치ficas a las cuales se accede a trav칠s de los 칤ndices.

````python
# Creamos los datos que vamos a utilizar
x = np.linspace(0,5,11)
y = x**2
z = y/2
````

````python
# Para crear una gr치fica que sea tipo Fig. necesitas trabajar el siguiente c칩digo
fig, axes = plt.subplots(1,2)
axes[0].plot(x,y)
axes[1].pie(x)
plt.show()
````

````python
fig, (axes1, axes2) = plt.subplots(1,2)
axes1.scatter(x,y)
axes2.bar(x,y)
plt.show()
````

````python
fig, axes = plt.subplots(2,4)
axes[0,0].stem(x,y)
axes[0,1].step(x,y)
axes[0,2].pie(x)
axes[0,3].fill_between(y,x)
fig.tight_layout()
````

````python
# Alternativa para generar una matriz de tipo fig, accediendo a cada gr치fica mediante los nombres dados en una tupla
fig, ((axes1, axes2, axes3, axes4), (axes5, axes6, axes7, axes8)) = plt.subplots(2,4)
axes1.plot(x, np.sin(x), 'b')
axes2.plot(y, np.cos(x), 'purple')
axes3.plot(x, np.tan(y), 'orange')
axes4.plot(y, np.cos(y)**2,'black')

#mejorar la visualizaci칩n de las figuras
fig.tight_layout()
````

````python
fig, axes = plt.subplots(1,3)
axes[0].stem(z,x)
axes[1].step(x,y)
axes[2].fill_between(x,y, color='skyblue')
plt.tight_layout()
````

### Leyendas, etiquetas, t칤tulos, tama침o

### Set

Set es un m칠todo de Seaborn que permite configurar el estilo, fuente y color de las gr치ficas.

````python
# Importamos librerias
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
````

**쯈u칠 necesitas saber al usar Set?**

### Jointplot y Pairplot

En resumen, jointplot es ideal para explorar la relaci칩n entre dos variables, mientras que pairplot es m치s adecuado para explorar relaciones y distribuciones entre m칰ltiples variables num칠ricas en un conjunto de datos. Ambas funciones son herramientas valiosas en el kit de herramientas de visualizaci칩n de datos de Seaborn, y la elecci칩n entre ellas depender치 de los objetivos espec칤ficos de la visualizaci칩n.

#### Jointplot

````python
# Joinplot une 2 distintas gr치ficas en una sola
sns.jointplot(x='total_bill', y='tip', data= tips, hue='sex');
````

````python
# Agregar m치s argumentos para poder hacer un mejor an치lisis
sns.jointplot(data=tips, x='total_bill', y= 'tip', hue='sex', kind= 'hist', marginal_ticks=True, #muestra una peque침a tabla para la gr치fica exterior
marginal_kws=dict(bins= 25, fill = True, multiple= 'dodge')); #marginal_kws se le pueden agregar diferentes par치metros
````

#### Pairplot

````python
# Pairplot correlaciona todas las variables num칠ricas que hay en el dataset
sns.pairplot(data=tips);
````

````python
# Solo muestra una esquina del pairplot, eliminando as칤, las gr치ficas repetidas
sns.pairplot(data= tips,hue='sex' , corner=True);
````

### Heatmap

Heatmap es un tipo de gr치fico enfocado a una estructura matricial. Heatmap correlaciona todas las variables num칠ricas del dataset.

````python
# Correlaci칩n entre variables
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

# Elimine las columnas no num칠ricas antes de calcular la correlaci칩n
numeric_tips = tips.select_dtypes(include=['float64', 'int64'])
numeric_tips.corr()
# Muestra las variables correlacionadas entre si
````

````python
# Heatmap
sns.heatmap(numeric_tips.corr());
````

````python
# Se pueden agregar diferentes par치metros. annot muestra el valor de la correlaci칩n,
# cmap color,linewidthsespacio entre variables, linecolor color de las l칤neas, vminv, max valores m치ximos y m칤nimos, cbar=False eliminar la barra
sns.heatmap(numeric_tips.corr(), annot= True, cmap='coolwarm', linewidths=5, linecolor='black', vmin=0.5,vmax=1,cbar=True);
````
